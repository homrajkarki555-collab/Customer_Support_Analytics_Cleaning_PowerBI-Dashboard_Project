```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('/content/Call Centre Dataset - Sheet1.csv')
```

```python
df.info()
```

```python
df['Unnamed: 0'].unique()
```

```python

# Define the renaming mapping
rename_map = {
    'Unnamed: 0': 'Conversation_ID',
    'Agent': 'Agent_name',
    'Speed of Answer': 'Respond_time',
    'AvgTalkDuration': 'Avg_handeling_time'
}

# Apply the rename
df.rename(columns=rename_map, inplace=True)

# Save the transformed data
df.to_csv('transformed_data.csv', index=False)

# Display the new column names
print(df.columns)
```

```python
df['Conversation_ID'].count()
```

```python
#  Ensure there are no missing values in Conversation_ID
df = df.dropna(subset=['Conversation_ID'])

#  Set Conversation_ID as the index (the DataFrame's version of a Primary Key)
df.set_index('Conversation_ID', inplace=True)

```

```python
df
```

```python
def time_to_seconds(time_str):
    if pd.isna(time_str):
        return 0
    # Split the time string by ':'
    parts = str(time_str).split(':')
    if len(parts) == 3: # Format HH:MM:SS
        return int(parts[0]) * 3600 + int(parts[1]) * 60 + int(parts[2])
    elif len(parts) == 2: # Format MM:SS
        return int(parts[0]) * 60 + int(parts[1])
    return 0

# Apply the transformation
df['Avg_handeling_time'] = df['Avg_handeling_time'].apply(time_to_seconds)
```

```python
df
```

```python
df['Agent_name'].unique()
```

```python
# 1. Treat 0 as missing for the handling time calculation
df['Avg_handeling_time'] = df['Avg_handeling_time'].replace(0, np.nan)

# 2. Define a function to fill missing values with the group (agent) mean
def fill_agent_averages(group):
    group['Respond_time'] = group['Respond_time'].fillna(group['Respond_time'].mean())
    group['Avg_handeling_time'] = group['Avg_handeling_time'].fillna(group['Avg_handeling_time'].mean())
    return group

# 3. Apply the function to each agent group
df = df.groupby('Agent_name', group_keys=False).apply(fill_agent_averages)

# 4. Global fallback for agents with zero valid entries
df['Respond_time'] = df['Respond_time'].fillna(df['Respond_time'].mean())
df['Avg_handeling_time'] = df['Avg_handeling_time'].fillna(df['Avg_handeling_time'].mean())
```

```python
df
```

```python
# Rounding Respond_time to  decimal places
df['Respond_time'] = df['Respond_time'].round(0)

# Rounding Avg_handeling_time to 0 decimal places (to keep seconds as whole numbers)
df['Avg_handeling_time'] = df['Avg_handeling_time'].round(0).astype(int)
```

```python
df
```

```python
df['Satisfaction rating'] = df['Satisfaction rating'].fillna(0) #Fill Null values in "Satisfaction rating" with 0
```

```python
#To verify it worked:
print(df['Satisfaction rating'].isnull().sum()) # This should print 0
```

```python
# Save the final DataFrame to a CSV file on your computer
df.to_csv('/content/Call Centre Dataset - Sheet1.csv', index=True)

print("File has been saved successfully!")
```
